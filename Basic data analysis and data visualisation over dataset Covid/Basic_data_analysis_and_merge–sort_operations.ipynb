# **Marge Operation**

```
dataset1 = pd.DataFrame([['elephant','vegetables'],['tiger','meat'],['kangaroo','vegetables'],['zebra','vegetables'],['giraffe','vegetables']],columns=['animal','food'])
```

```
dataset1
```

```
dataset.merge(dataset1)
```

### Pandas Mergeâ€¦ But how? Inner, outer, left or right?

```
dataset.merge(dataset1, how='outer')
```

```
dataset.merge(dataset1, how = 'left')
```

```
dataset.merge(dataset1, how = 'left', left_on = 'animal', right_on = 'animal')
```

# **sorting operations**

```
dataset.sort_values('water_need')
```

```
dataset.sort_values(by = ['animal', 'water_need'])
```

```
dataset.sort_values(by = ['uniq_id'], ascending = False)
```

## **Reset_index**

```
dataset.sort_values(by = ['water_need'], ascending = False).reset_index()
```

## **Fillna**

```
dataset.merge(dataset1, how = 'left')
```

```
dataset.merge(dataset1, how = 'left').fillna('unknown')
```

```
dataset.merge(dataset1, how='left').fillna('MEAT')
```
